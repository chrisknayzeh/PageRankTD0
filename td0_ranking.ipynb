## Libraries used

```python
#Libraries
import numpy as np
import time
```

# Functions Created

```python
#Function that reads a text file and returns the lines in a list, the number of rows and columns (in this case, row = col)

def readText(file):
  text = open(file)
  lines = text.readlines()
  row = int(lines[0])
  col = int(lines[0])

  return lines, row, col
```

```python
#Function to print a matrix in a nice way (literally for convenience)

def printMat(matrix):
  rows = []
  for i in range(row):
    rows.append(matrix[i])
  
  for i in range(row):
    print(rows[i] , "\n")
```

```python
#Function to find vector f for a graph
def findF(lines):
  N = len(lines)
  f = list()
  for i in range(N):
    if (lines[i][2] == 0):
      f.append(1)
    else:
      f.append(0)
  return f
```

```python
#Function to Remove spaces and new lines from lines

def rmRed(lines):
  newlines = list(lines[i] for i in range(2,len(lines)))
  #Removing \n from the strings
  for i in range (2,len(lines)):
    lines[i]=lines[i].replace('\n',"")
  #Using the " " as a delimiter for new elements. Each number will become an element on its own, and newlines will become a list of lists
  for i in range(len(newlines)):
    newlines[i] = newlines[i].split(" ")
  return list(newlines)
```

```python
#Function to convert string to numeric

def convertStrNum(newlines):
  for i in range(len(newlines)):
    for j in range(len(newlines[i])):
      if(j<3 or j%2==0): #since we start indexing at 0, and we are interested in the values starting the third element, any values before the third element will become int.
                       #Also if the value is on an even index (0, 2, 4, 8 etc.) it will also be an int since it's either the row number, number of edges or a vertex number.
                       newlines[i][j] = int(newlines[i][j])
      else:
        newlines[i][j] = float(newlines[i][j]) #Otherwise, the rest of the values are the probabilities (or weights)
  return newlines
```

# Execution
All prints are in separated blocks in this code. They are purely to check the results and not essential to the code working properly.

To print a list of lists line by line, use ***printMat(list)*** instead of ***print(list)***

```python
#Reading the file.
#file name is "graphe_a_8_sommet.txt", dont forget to import the file

lines, row, col = readText("graphe_a_8_sommet.txt")
```

```python
print(lines)
print(row)
print(col)
```

```python
#Remove spaces and new lines from lines
newlines = rmRed(lines)

newlines = convertStrNum(newlines)
```

```python
printMat(newlines)
printMat(lines)
```

```python
#Creating an empty matrix
matrix = [[0 for i in range(row)] for j in range(col)]

#Creating the stochastic matrix of the graph

for i in range(row):
  j=2
  while j < len(newlines[i]):
    matrix[i][newlines[i][j]-1] = newlines[i][j+1]
    j+=2
```

```python
printMat(matrix)
```

# Matrix Product

```python
#Creating a numpy array from matrix, vector from the given and executing the matrix product v*M
M = np.array(matrix)
vector = np.array([0.1,0.1,0.2,0,0.1,0.15,0.15,0.2]) #initialization of the vector

result = np.matmul(vector,M)
```

```python
print(result)
```

```python
#Recursive loop until the product converges
e = 10**(-6)
i=1
start = time.time()
while(abs(vector-result) > e).any():
  vector = result
  result = np.matmul(vector,M)
  #print("\nIteration ",i,"\n")
  #print("Vector", vector)
  #print("Result",result)
  i=i+1
print("End.\nIterations: ", i-1,"\nRuntime: ", time.time()-start," seconds")
```

G = alphaP + alpha transpose(f) . e/N + (1-alpha) transpose(e).e/N

```python
#Read new graph (101 vertices)
lines, row, col = readText("graph101.txt")

N = row
```

```python
printMat(lines)
```

```python
#Creating an empty matrix
matrix = [[0 for i in range(row)] for j in range(col)]
e = [1 for i in range(row)]

#Creating the stochastic matrix of the graph

for i in range(row):
  j=2
  while j < len(lines[i]):
    matrix[i][lines[i][j]-1] = lines[i][j+1]
    j+=2
```

```python
printMat(matrix)
```

```python
#Creating a numpy array from matrix, vector from the given and executing the matrix product v*M
M = np.array(matrix)
vector = [1/N for i in range(row)] #initialization of the vector
vector = np.array(vector)
```

```python
result = np.matmul(vector,M)
#Recursive loop until the product converges
e = 10**(-8)
i=1
start = time.time()
while(abs(vector-result) > e).any():
  vector = result
  result = np.matmul(vector,M)
  #print("\nIteration ",i,"\n")
  #print("Vector", vector)
  #print("Result",result)
  i=i+1
print("End.\nIterations: ", i-1,"\nRuntime: ", time.time()-start," seconds")
```

```python
f101 = findF(lines)
f101=np.matrix(f101)
alpha = 0.85
#print(f101)
```

```python
elt = np.array([1 for i in range(row)])
xP = np.array(result)
v = np.array(vector)
ft = np.transpose(f101)
```

```python
xG = alpha * result + alpha*(v*ft)*elt/N + (1-alpha)*elt/N

print(xG)
```
